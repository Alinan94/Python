```python
#how to use beautiful soup webscraper
from bs4 import BeautifulSoup
import requests
```

```python
url = input ('Enter URL: ')
```

```python
#Uses request data to get http data
r = requests.get('http://' + url)
```

```python
#get response
data = r.text
```

```python
#BeautifulSoup function takes one argument and an uses its own html parser
soup = BeautifulSoup(data, 'html.parser')
```

```python
print(soup.get_text())
```

```python
#How to extract hyperlinks
for link in soup.find_all ('a'):
    print(link.get('href'))
```

```python
#How to extract Images
for image in soup.find_all('img'):
    print(image.get('src'))
```
