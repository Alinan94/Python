```python
#matplot lib helps visualize data in form of graphs
import matplotlib.pyplot as plt
from matplotlib import style
```

```python
#uses classic style graph
style.use('classic')
#graphs category of max temp
df['Min.TemperatureC'].plot()
#shows graph
plt.show()
```

```python
#in pandas, NaN means there is a missing value
# this statment says that all NaN data will be replaced with the mean value of Max Temp
df['Max.TemperatureC'].fillna (df['Max.TemperatureC'].mean(), inplace = True)
```
